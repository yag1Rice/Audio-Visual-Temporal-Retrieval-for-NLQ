```python
from pathlib import Path

folder = Path("/scratch/yag1/ego4d_data/v2/clips")

subdirs = [p.name for p in folder.iterdir() if p.is_dir()]
print(subdirs)
```

```python
import shutil
```

```python
for subdir in subdirs:
    video_embeddings_dir = folder / subdir / "video_embeddings"
    audio_embeddings_dir = folder / subdir / "audio_embeddings"
    shutil.rmtree(video_embeddings_dir)
    shutil.rmtree(audio_embeddings_dir)
```
