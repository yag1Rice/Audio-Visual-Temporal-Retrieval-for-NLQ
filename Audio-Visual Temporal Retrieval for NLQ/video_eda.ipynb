```python
from pathlib import Path

dir_string = "/scratch/yag1/ego4d_data/v2/clips"
dir_path = Path(dir_string)
video_paths = [p.name for p in dir_path.iterdir() if p.is_file() and p.suffix == ".mp4"]
```

```python
video_paths_with_audio = []
video_paths_without_audio = []

for path in video_paths:
    video = VideoFileClip(dir_string + "/" + path)

    if video.audio is not None:
        print(f"{path}: HAS audio")
        video_paths_with_audio.append(path)
    else:
        print(f"{path}: NO audio")
        video_paths_without_audio.append(path)

    video.close()
```

```python
print(len(video_paths_with_audio))
print(len(video_paths_without_audio))
```
